## `Produtos`

```python
# Carregar dataframe
from joblib import load
df_produtos = load("../../../util/data/df_produtos.z")
```

Iniciaremos observando as informações do dataframe e se é necessário uma análise de possíveis outliers:

```python
df_produtos.info()
```

```python
df_produtos.describe()
```

```python
df_produtos.head()
```

Nesse cenário, iniciaremos a análise da coluna `preco_venda`.

O primeiro passo é visualizar a distribuição da variável `preco_venda` com um histograma e um boxplot. Isso ajuda a entender a forma dos dados e a identificar possíveis padrões ou anomalias:

```python
import matplotlib.pyplot as plt
import seaborn as sns

'''
# Histograma
plt.figure(figsize=(10, 6))
sns.histplot(df_produtos['preco_venda'], kde=True, bins=50, color='blue')
plt.title('Distribuição de Preço de Venda')
plt.xlabel('Preço de Venda')
plt.ylabel('Frequência')
plt.show()
'''


# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_produtos['preco_venda'], color='orange')
plt.title('Boxplot de Preço de Venda')
plt.xlabel('Preço de Venda')
plt.show()
```

Em seguida utilizaremos o Z-score para identificar outliers ao medir o número de desvios padrão que um valor está distante da média.

```python
from scipy.stats import zscore

# Calcular Z-score
df_produtos['zscore'] = zscore(df_produtos['preco_venda'])

# Identificar outliers com Z-score > 3 ou < -3
outliers_zscore = df_produtos[df_produtos['zscore'].abs() > 3]

print(f'Outliers identificados pelo Z-score:\n{outliers_zscore}')
```

Com isso, será necessário escolher uma estratégia para lidar com esses possíveis outliers.

*Comentário: Embora sejam detectados como outliers, aparentemente são valores que fazem sentido na regra de negócio.
